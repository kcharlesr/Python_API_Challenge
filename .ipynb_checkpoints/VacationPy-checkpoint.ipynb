```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
file = 'Food_Production.csv'
df = pd.read_csv(file)
```

```python
#df.columns
food_df = df[['Food product', 'Land use per 1000kcal (m² per 1000kcal)']]
```

```python
food_df
```

```python
drop_df = food_df.dropna()
drop_df
```

```python
new_index_df = drop_df.reset_index(drop = True)
new_index_df
```

```python
plt.figure(figsize=(15,10))
plt.bar(x=new_index_df['Food product'], height=new_index_df['Land use per 1000kcal (m² per 1000kcal)'])
plt.xticks(rotation = 90)
plt.show()
```
